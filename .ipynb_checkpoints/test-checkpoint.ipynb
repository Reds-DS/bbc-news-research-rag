```python
import pandas as pd
import os
import weaviate
from sentence_transformers import SentenceTransformer
from dotenv import load_dotenv
import weaviate.classes as wvc
```

```python
## Import Data
df = pd.read_csv("Data/bbc_news.csv")
```

```python
load_dotenv()

hf_api_key = os.getenv("HUGGINGFACE_API_KEY")

## Connect to weaviate
client = weaviate.connect_to_local(
    host="weaviate",
    port=8080,
    grpc_port=50051
)

print("Connected:", client.is_ready())
```

```python
# Check if collection exists
if client.collections.exists("BBC_NEWS"):
    client.collections.delete("BBC_NEWS")

# Create data collection
bbc_news = client.collections.create(
    name="BBC_NEWS",
    vector_config=wvc.config.Configure.Vectors.text2vec_huggingface(
        model='sentence-transformers/all-MiniLM-L6-v2'
    ),
    properties=[
        Property(name="title", data_type=DataType.TEXT),
        Property(name="pubDate", data_type=DataType.TEXT),
        Property(name="guid", data_type=DataType.TEXT),
        Property(name="link", data_type=DataType.TEXT),
        Property(name="description", data_type=DataType.TEXT)
    ]
)
```

```python
# Import data objects
bbc_news_objects = list()
for _, row in df.iterrows():
    properties = {
        "title": row.title,
        "pubDate": row.pubDate,
        "guid": row.guid,
        "link": row.link,
        "description": row.description,
    }
    bbc_news_objects.append(properties)

# bbc_news.data.insert_many(bbc_news_objects)
BATCH_SIZE = 100

for i in range(0, len(bbc_news_objects), BATCH_SIZE):
    bbc_news.data.insert_many(bbc_news_objects[i:i + BATCH_SIZE])
```
